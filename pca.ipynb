#Dimensionality Reduction with PCA

#IMPORTS

```python
#Import Numpy And Pandas
import numpy as np
import pandas as pd
```

```python
#Import Matplotlib And Seaborn
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

#GET THE DATA

```python
# Import the breast cancer dataset from scikit-learn
from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer()
```

```python
#Display the names of the features in the breast cancer dataset
cancer.feature_names
```

```python
#Create a DataFrame using the feature data from the breast cancer dataset, with column names set to feature names.
df = pd.DataFrame(data=cancer.data,columns=cancer.feature_names)
```

```python
#Head Of df
df.head()
```

```python
# Display the target labels  in the breast cancer dataset
cancer.target
```

```python
#Display the names of the target classes (['malignant', 'benign']) in the breast cancer dataset
cancer.target_names
```

#SCALING FEATURES

```python
# Import StandardScaler to normalize feature
from sklearn.preprocessing import StandardScaler
```

```python
#Create an instance of StandardScaler to standardize the dataset
scaler = StandardScaler()
```

```python
#Fit Scaler On df
scaler.fit(df)
```

```python
#Transform our data
scaled_features = scaler.transform(df)
```

```python
scaled_features
```

#PCA

```python
# Import PCA (Principal Component Analysis) for dimensionality reduction
from sklearn.decomposition import PCA
```

```python
# Create a PCA instance to reduce the dataset to 2 principal components
pca = PCA(n_components=2)
```

```python
#Compute the principal components from the standardized feature set
pca.fit(scaled_features)
```

```python
#Apply the trained PCA transformation to reduce dimensionality of the dataset
x_pca=pca.transform(scaled_features)
```

```python
print(x_pca.shape)
print('\n')
print(scaled_features.shape)
```

```python
# Create a new DataFrame to store the principal components  
first_component = []
second_component = []
for i in range(569):
    first_component.append(x_pca[i][0])
    second_component.append(x_pca[i][1])
new_df = pd.DataFrame()
new_df['First Component']=first_component
new_df['Second Component']=second_component
```

```python
new_df
```

```python
# Scatter plot to visualize data points in the reduced 2D space after PCA transformation
plt.scatter(new_df['First Component'], new_df['Second Component'], alpha=0.5, c=cancer.target, cmap='coolwarm')  
plt.xlabel('First Principal Component')  
plt.ylabel('Second Principal Component')  
plt.title('PCA Projection of Breast Cancer Data')  
plt.colorbar(label='Target (0 = Malignant, 1 = Benign)')  
plt.show()
```

```python
#Components
pca.components_
```

```python
# Create a DataFrame to visualize how much each original feature contributes to the principal components
comp = pd.DataFrame(pca.components_,columns=cancer.feature_names)
```

```python
comp
```

```python
#Create a heatmap to visualize the contribution of each original feature to the principal components
sns.heatmap(comp,cmap='plasma')
```
